```python
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Downloads/mylci_datadownload_20230222_234730.csv')
```

```python
df.head()
```

# Q 1: How many Regions exists and membership distribution in each region?

```python
print('Number of Regions: ', df['Region Name'].nunique())
print(df['Region Name'].value_counts())
df['Region Name'].value_counts().plot(kind = 'bar')
```

```python
df['Region Name'].value_counts().plot(kind = 'pie', autopct = '%2.f')
```

# Q 2: How many zones exist and membership distribution in each zone?

```python
print('Number of Zones: ', df['Zone Name'].nunique())
print(df['Zone Name'].value_counts())
df['Zone Name'].value_counts().plot(kind = 'bar', color = 'blue')
```

```python
df['Zone Name'].value_counts().plot(kind = 'pie', autopct = '%2.f')
```

# Q 3: Top 10 Clubs according to Membership?

# Q 5: Top 10 Cities with respect to membership?

```python
print("Number of Cities members are from: ", df['Member Address City'].nunique())
print("Top 10 cities with maximum membership: ")
df['Member Address City'].value_counts().head(10)
```

# Q 6: Gender based Membership Distribution?

```python
print('Total Members in District 305-N2: ', df['Gender'].count())
print(df['Gender'].value_counts())
df['Gender'].value_counts().plot(kind = 'pie', autopct = '%2.f')
```

# Q 7: Members Age Distribution?

```python
df['Date of Birth'] = pd.to_datetime(df['Date of Birth'])

current_year = dt.datetime.now().year

df['Age'] = current_year - df['Date of Birth'].dt.year
```

```python
df['Age'].plot(kind='hist', bins=20, figsize=(10,5), xlabel='Age', ylabel='Number of Members', 
               title='Age Distribution of Members')
```

```python
# create a figure and axis
fig, ax = plt.subplots(figsize=(10, 5))

# plot the histograms by gender
df.hist(column='Age', by='Gender', bins=20, ax=ax, edgecolor='black')

# set the x-axis label
ax.set_xlabel('Age')

# set the y-axis label
ax.set_ylabel('Number of Members')

# add a title
ax.set_title('Age Distribution of Members by Gender')

# show the plot
plt.show()
```

# Q 8: Tenure of a Member?

```python
df['Join Date'] = pd.to_datetime(df['Join Date'])

current_date = dt.datetime.now().year

df['Join Year'] = current_date - df['Join Date'].dt.year
```

```python
df['Join Year'].value_counts().plot(kind = 'bar', color = 'blue')
```
